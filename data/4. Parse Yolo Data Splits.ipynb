Randomly split between train and val for images we have annotated (and blanks)

```python
print('Copying files for yolo splits.')

        
#Convert image files in directory to event numbers
image_files = [f for f in listdir(toppath + 'yolo_images')]
   
image_numbers = []

for image in image_files:
    if 'SSWI' in image:
        image_numbers.append(image[:-5])
    else:
        image_numbers.append(image.split('_')[0])
        
image_numbers = set(image_numbers)    

for number in image_numbers:

    #Determine folder to move images, 80,20 split
    rand_num = random.uniform(0, 1)
    if rand_num < .8:
        split = 'train/'
    else:
        split = 'val/'

        
    if 'SSWI' in number:
        image_appendix = ['A.jpg', 'B.jpg', 'C.jpg']
        
    else:
        image_appendix = ['_0.jpeg', '_1.jpeg', '_2.jpeg']
        
    for image in image_appendix:
        try:
            shutil.copy(toppath + 'yolo_images/' + number + image,
                        toppath + 'yolo_splits3/' + split + 'images')
            try:
                shutil.copy(toppath + 'yolo_extract/' + number + image.split('.')[0] + '.txt',
                        toppath + 'yolo_splits3/' + split + 'labels')
            except:
                print('issue with extract' + number + image)
        except:
            continue

print('Files for yolo model copied.')


```
