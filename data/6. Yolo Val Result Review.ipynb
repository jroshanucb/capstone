```python
#Species model results

event_id_list = []
image_id_list = []
true_class_list = []
pred_class_list = []
pred_class_score_list = []
true_count_list = []
pred_count_list = []

for i in range(len(data['data'])):
    row = data['data'][i]
    
    #Event ID
    event_id = row[2]['path'].split('data/images/')[1].split('_')[0]
    
    #Image ID
    image_id = row[2]['path'].split('data/images/')[1].split('.jpeg')[0]

    #True Count
    true_count = len(row[2]['boxes']['ground_truth'])

    #True class    
    if true_count == 0:
        true_class = 'blank'
    else:

        true_class = row[2]['boxes']['ground_truth'][0]['box_caption']

    #Pred Count
    pred_count = len(row[3]['boxes']['predictions'])

    #Pred class
    if pred_count == 0:
        pred_class = 'blank'
        pred_class_score = 0

    else:
        pred_score_for_image_list = []
        for score in range(0, len(row[3]['boxes']['predictions'])):
            int_score = row[3]['boxes']['predictions'][score]['scores']['class_score']

            pred_score_for_image_list.append(int_score)
        #index of max score of predictions
        index_max = max(range(len(pred_score_for_image_list)), key=pred_score_for_image_list.__getitem__)
        print(pred_score_for_image_list)
        pred_class = row[3]['boxes']['predictions'][index_max]['box_caption'].split(' ')[0]
        pred_class_score = row[3]['boxes']['predictions'][index_max]['box_caption'].split(' ')[1]

        
        
    event_id_list.append(event_id)
    image_id_list.append(image_id)
    true_class_list.append(true_class)
    pred_class_list.append(pred_class)
    pred_class_score_list.append(pred_class_score)
    true_count_list.append(true_count)
    pred_count_list.append(pred_count)
    
    
val_result_df['event_id'] = event_id_list
val_result_df['image_id'] = image_id_list
val_result_df['true_class'] = true_class_list
val_result_df['pred_class'] = pred_class_list
val_result_df['pred_class_score'] = pred_class_score_list
val_result_df['true_count'] = true_count_list
val_result_df['pred_count'] = pred_count_list

    
```
